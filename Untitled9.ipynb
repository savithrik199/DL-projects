```
code={"benign":0,"malignant":1}
```

```
x_train = []
y_train = []
for folder in  os.listdir(train) : 
    files = gb.glob(pathname= str( train +'/' + folder + '/*.jpg'))
    for file in files: 
        images = cv2.imread(file)
        x_train.append(list(images))
        y_train.append(code[folder])
print(f'we have {len(x_train)} items in X_train')
print(f'we have {len(y_train)} items in y_train')
```

```
x_test = []
y_test = []
for folder in  os.listdir(test) : 
    files = gb.glob(pathname= str( test +'/' + folder + '/*.jpg'))
    for file in files: 
        images = cv2.imread(file)
        x_test.append(list(images))
        y_test.append(code[folder])
print(f'we have {len(x_test)} items in X_test')
print(f'we have {len(y_test)} items in y_test')
```

```
import numpy as np

x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```
x_train.shape
```

```
x_train=x_train/255
```

```
x_test=x_test/255
```

```

    
model1=tf.keras.applications.ResNet50V2(include_top=True,weights='imagenet',input_shape=(224,224,3))
    
```

```
model1.summary()
```

```
for la in model1.layers:
  la.trainable=False
```

```
    rein=model1.layers[0].input
    reout=model1.layers[-2].output
    predictions=tf.keras.layers.Dense(2)(reout)
```

```
model1.summary()
```

```
model2=keras.Model(inputs=rein,outputs=predictions)
```

```
model2.summary()
```

```
model2.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```
model2.fit(x_train,y_train,epochs=10)
```

```
model2.evaluate(x_test,y_test)
```
