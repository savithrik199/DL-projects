```
code = {'buildings':0 ,'forest':1,'glacier':2,'mountain':3,'sea':4,'street':5}

def getcode(n) : 
    for x , y in code.items() : 
        if n == y : 
            return x
```

```
s=100
X_train = []
y_train = []
for folder in  os.listdir(train) : 
    files = gb.glob(pathname= str( train + folder + '/*.jpg'))
    for file in files: 
        image = cv2.imread(file)
        image_array = cv2.resize(image , (s,s))
        X_train.append(list(image_array))
        y_train.append(code[folder])
```

```
X_test = []
y_test = []
for folder in  os.listdir(test) : 
    files = gb.glob(pathname= str(test+ folder + '/*.jpg'))
    for file in files: 
        image = cv2.imread(file)
        image_array = cv2.resize(image , (s,s))
        X_test.append(list(image_array))
        y_test.append(code[folder])
```

```
len(X_test)
```

```
len(y_test)
```

```
X_pred = []
files = gb.glob(pathname= str(pred+ '/*.jpg'))
for file in files: 
    image = cv2.imread(file)
    image_array = cv2.resize(image , (s,s))
    X_pred.append(list(image_array))       
```

```
len(X_pred)
```

```
X_train = np.array(X_train)
X_test = np.array(X_test)
X_pred = np.array(X_pred)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```
intel = keras.models.Sequential([
        layers.Conv2D(512,kernel_size=(3,3),activation='relu',input_shape=(s,s,3)),
        layers.Conv2D(256,kernel_size=(3,3),activation='relu'),
        layers.MaxPool2D(3,3),
        layers.Conv2D(256,kernel_size=(3,3),activation='relu'),    
        layers.Conv2D(128,kernel_size=(3,3),activation='relu'),
        layers.MaxPool2D(3,3),
        layers.Flatten() , 
        layers.Dropout(rate=0.4),
        layers.BatchNormalization(), 
        layers.Dense(200,activation='relu') , 
        layers.Dropout(rate=0.3),
        layers.BatchNormalization(),  
        layers.Dense(150,activation='relu') ,            
        layers.Dropout(rate=0.5) ,            
        layers.Dense(6,activation='softmax') ,    
        ])
```

```
intel.compile(optimizer ='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```
intel.summary()
```

```

intel.fit(X_train, y_train, epochs=20,batch_size=100)
```

```
 intel.evaluate(X_test, y_test)
```

```
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import pandas as pd
import seaborn as sn
```

```
y_pre = intel.predict(X_test)
y_predicted_labels = [np.argmax(i) for i in y_pre]
```

```
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```
